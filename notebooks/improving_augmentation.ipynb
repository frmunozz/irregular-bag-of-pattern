```python
detections = {}
for obj in train_set.objects:
    c = np.sum(obj.observations["detected"].to_numpy())
    detections[obj.metadata["object_id"]] = c
```

```python
# count objects with more than N detections
def detections_threshold(detections, N):
    c = 0
    det2 = {}
    dropped = {}
    for k, v in detections.items():
        if v > N:
            c+=1
            det2[k] = v
        else:
            dropped[k] = v
    print(c)
    return det2, dropped

detections2, dropped = detections_threshold(detections, 2)
```

```python
def count_augment_per_object(df, dropped):
    reference_df = df[df["reference_object_id"].isnull()]
    df2 = df[~df["reference_object_id"].isin(dropped)]
    
    augment_count = {}
    for obj_id in reference_df.index:
        df_i = df2[df2["reference_object_id"] == obj_id]
        augment_count[obj_id] = len(df_i)
    
    return augment_count

augment_count = count_augment_per_object(df, dropped)
```
