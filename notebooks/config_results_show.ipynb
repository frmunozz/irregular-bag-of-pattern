```python
result_path3 = os.path.join(main_path, 'data', 'configs_results', "multi_ress_search")
files3 = glob.glob(os.path.join(result_path3, "*"))
files3
```

```python
def get_stack(f):
    base, name = os.path.split(f)
    stack = name.split("--")[1].split("-")[0]
    return stack

def filter_files(files, config, top_n):
    files2 = []
    for f in files:
        if config in f:
            if "top_%d" % top_n in f:
                files2.append(f)
    return files2
```

```python
# options = ["alph_size_6"]
plt.figure(figsize=(16, 5))
wl = 1
config = "multi_ress-MTm"
top_n = 1
top_0_file = None
for f in files3:
    if "top_0" in f:
        if config in f:
            top_0_file = f
            print(f)
        
df = pd.read_csv(top_0_file)
df = df[df["scheme"] == "ltc"]
df = df[df["wl"] == wl]
plt.errorbar(df["win"], df["mean_cv"], label="stack_1")

# wl = 1
# print(len(options))
# main_df = None
# for f in files2:
#     q = get_alph(f)
#     if q in options:
#         print(q, f)
#         df = pd.read_csv(f)
#         df = df[df["scheme"] == "ltc"]
# #         df = df[df["wl"] == wl]
#         plt.errorbar(df["win"], df["mean_cv"], yerr=df["std_cv"], label="stack_1")
    
options = ["stack_2", "stack_3", "stack_4", "stack_5"]
# plt.figure(figsize=(16, 5))
print(len(options))
main_df = None
highest_acc_config = None
highest_acc = -1
files_filtered = filter_files(files3, config, top_n)
for f in np.sort(files_filtered):
    q = get_stack(f)
    if q in options:
        print(q, f)
        df = pd.read_csv(f, index_col=None)
        df = df[df["scheme"] == "ltc"]
        best_acc = np.max(df["mean_cv"])
        if best_acc > highest_acc:
            idx = np.argmax(df["mean_cv"])
            highest_acc_config = ", ".join([", ".join(df.index[0]), df["base"][0]]) + ", {%f, %d}" % (df["win"][idx], df["wl"][idx])
            highest_acc = best_acc
        df = df[df["wl"] == wl]
        df = df.sort_values(by=["win"])
        if "stack_5" in f:
            plt.errorbar(df["win"], df["mean_cv"], yerr=df["std_cv"], label=q)
        else:
            plt.errorbar(df["win"], df["mean_cv"], label=q)

plt.xscale("log")
plt.legend()
print(highest_acc_config, highest_acc)
```
